```python
import datetime, time
import simpy

import pandas as pd

import openclsim.core as core
import openclsim.model as model
import openclsim.plot as plot
```

# Start at timeout

```python
simulation_start = 0
my_env = simpy.Environment(initial_time=simulation_start)
registry = {}

activity = model.BasicActivity(
    env=my_env,
    name="Basic activity",
    registry=registry,
    duration=10,
    start_event={
        "type":"time",
        "start_time":10,
    }
    
)

model.register_processes([activity])
my_env.run()
```

```python
plot.get_log_dataframe(activity, [activity])
```

# Repeat until timeout

```python
simulation_start = 0
my_env = simpy.Environment(initial_time=simulation_start)
registry = {}


activity = model.WhileActivity(
    env=my_env,
    name="While activity of basic activities",
    registry=registry,
    sub_processes=[
    model.BasicActivity(
            env=my_env,
            name="Basic activity",
            registry=registry,
            duration=10,    
        )
    ],
    condition_event={
        "type":"time",
        "start_time":50,
    },
)


model.register_processes([activity])
my_env.run()
```

```python
plot.get_log_dataframe(activity, [activity])
```
