# CALSPEC HST Standard Star Spectra

```python
import os
import numpy as np
import scipy as sp
import scipy.io.idl
import astropy as ap
import astropy.units
import astropy.table
import astropy.io.fits
import synphot

import matplotlib.pyplot as plt

import lezargus

%matplotlib
```

```python
lezargus.initialize.initialize_logging_outputs()
# Making sure that the output directory exists.
os.makedirs("./products/standard_stars/", exist_ok=True)
```

### Configuration

```python
# The wavelength range in meters for the standard stars.
# The maximum wavelength is much longer, just because we have the WISE bands
# built in as well, and they don't seem to like to trim their wavelengths.
TRIM_WAVE_MINIMUM = (0.275 - 0.001) * 1e-6
TRIM_WAVE_MAXIMUM = (29.000 + 0.01) * 1e-6
```

```python
# We convert from flam units to the more typical SI units used in Lezargus.
stis_unit = astropy.units.Unit("erg s^-1 cm^-2 AA^-1")
lz_unit = astropy.units.Unit("W m^-2 m^-1")
conversion_factor = (lz_unit / stis_unit).decompose().scale
conversion_factor = stis_unit.to(lz_unit)  # Unknown issue.
```

# Sun

```python
sun_filename = "./base/standard_stars/sun_reference_stis_002.fits"

sun_fitsfile = astropy.io.fits.open(sun_filename)
sun_fitsmeta = sun_fitsfile[0].header
sun_fitsheader = sun_fitsfile[1].header
sun_fitsdata = sun_fitsfile[1].data

# The magnitudes for the Sun are from 2018ApJS..236...47W.
sun_header = {}
sun_header["LZO_NAME"] = "Sun"
sun_header["LZI_INST"] = "STScI CALSPEC"
sun_header["LZPM_J_U"] = -25.97
sun_header["LZPU_J_U"] = None
sun_header["LZPM_J_B"] = -26.13
sun_header["LZPU_J_B"] = None
sun_header["LZPM_J_V"] = -26.76
sun_header["LZPU_J_V"] = None
sun_header["LZPM_G_G"] = None
sun_header["LZPU_G_G"] = None
sun_header["LZPM_GGB"] = None
sun_header["LZPU_GGB"] = None
sun_header["LZPM_GGR"] = None
sun_header["LZPU_GGR"] = None
sun_header["LZPM_2_J"] = -27.90
sun_header["LZPU_2_J"] = None
sun_header["LZPM_2_H"] = -28.25
sun_header["LZPU_2_H"] = None
sun_header["LZPM_2Ks"] = -28.30
sun_header["LZPU_2Ks"] = None
```

```python
header_object = sun_header
data_object = sun_fitsdata

wave = data_object["WAVELENGTH"] / 1e10
flux = data_object["FLUX"] * conversion_factor
try:
    error = np.sqrt(
        data_object["STATERROR"] ** 2 + data_object["SYSERROR"] ** 2
    )
except Exception:
    error = np.sqrt(data_object["SYSERROR"] ** 2)

trim_index = np.where((TRIM_WAVE_MINIMUM <= wave) & (wave <= TRIM_WAVE_MAXIMUM))
trim_wave = np.array(wave[trim_index])
trim_flux = np.array(flux[trim_index])
trim_error = np.array(error[trim_index])

# We package the data into plain-text MRT tables. The actual meta data needs
# to be added later.
wavelength_column = astropy.table.Column(
    trim_wave,
    name="wavelength",
    unit="m",
    dtype=float,
)
flux_column = astropy.table.Column(
    trim_flux, name="flux", unit="W m^-2 m^-1", dtype=float
)
uncertainty_column = astropy.table.Column(
    trim_error, name="uncertainty", unit="W m^-2 m^-1", dtype=float
)

table_columns = [wavelength_column, flux_column, uncertainty_column]
libspec_table = astropy.table.Table(table_columns)
# Too much precision makes unnessary large file sizes.
libspec_filename = f"./products/standard_stars/star_spectra_Sun.dat"
libspec_table.write(
    libspec_filename,
    format="ascii.mrt",
    formats={keydex: "%.7e" for keydex in libspec_table.keys()},
    overwrite=True,
)
```

# Headers

```python
# The inital data, taken from the templates above.
star_header_dict = {
    "star_spectra_Sun": sun_header,
    "star_spectra_Vega": vega_header,
    "star_spectra_A0V": a0v_header,
    "star_spectra_109Vir": vir109_header,
    "star_spectra_16CygB": cyg16b_header,
}
# We pull every header key which might be added.
all_header_keys = []
for __, headerdex in star_header_dict.items():
    all_header_keys += list(headerdex.keys())
all_header_keys = sorted(set(all_header_keys))
```

```python
# We now create a table for the header keys.
star_header_columns = {}
# We start with the column keys.
star_header_columns["key"] = all_header_keys

# Adding the elements, in order.
for stardex, headerdex in star_header_dict.items():
    star_column = []
    for keydex in all_header_keys:
        star_column.append(headerdex.get(keydex, None))
    star_header_columns[stardex] = star_column

# Creating the table.
star_header_table = astropy.table.Table(star_header_columns)
# Writting the table.
star_header_filename = f"./products/standard_stars/star_header.dat"
star_header_table.write(
    star_header_filename, format="ascii.fixed_width", overwrite=True
)
```
