## Bending coupler

Bending coupler at the input side to reduce the side-reflection

1. Set the bending coupler height.
2. Guess the bending coupler length region.
3. Run the function to find the optimal bending coupler length.
4. You can use the optimal length in your `config_edge_coupler.yaml` settings.

```python
import numpy as np 
from gdslayout.utils import bending_coupler

h = 10
width=0.8
θ0 = np.pi * 22.5/180
resolution = 0.1

# guess the coupler length to be curved at height h and angle θ0, with width region [20, 60].
result = bending_coupler([20, 60], h, θ0, width, resolution)
result.draw_ports()
result.plot()
```
