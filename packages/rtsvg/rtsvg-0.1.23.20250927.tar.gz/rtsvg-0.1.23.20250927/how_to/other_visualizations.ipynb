# Other Visualizations
## Overview
Three other visualizations perform the same approximate function:  histograms, pie charts, and boxplots.
## Basics
Required parameters include:
- bin_by:  for the histogram and the boxplot
- color_by:  for the pie chart which is equivalent to the bin_by parameter in the other two visualizations

Commonly used parameters include:
- count_by:  determines the overall counting method applied to the plots -- if not specified (or set to 'None'), counting by rows will occur.

```python
import pandas as pd
import numpy as np
from rtsvg import *
rt = RACETrack()

counts = [1,    2,    3,    2,      10,    12,    12,    16,      6,       6,       6,          10]
colors = ['red','red','red','red',  'blue','blue','blue','blue',  'yellow','yellow','yellow',   'orange']
bins   = ['a',  'b',  'c',  'a',    'a',   'b',   'c',   'a',     'a',     'b',     'c',        'a']
df = pd.DataFrame({'counts':counts, 'colors':colors, 'bins':bins})
rt.tile([rt.histogram(df, bin_by='colors', color_by='colors', count_by='counts', h=80),
         rt.pieChart(df, color_by='colors', count_by='counts', w=80, h=80),
         rt.boxplot(df, bin_by='colors', color_by='colors', count_by='counts', style='barchart', h=80)])
```

## Boxplot
The boxplot component functions similarly to the boxplot configuration described in the temporal how-to.  Note that ordering can be specified with a number of different options to include user-specified order.  The "order_by" parameter can include the following values:
- 'sum'
- 'max'
- 'median'
- 'average' or 'mean'
- 'min'
- user-specified list

```python
rt.tile([rt.boxplot(df, bin_by='colors', color_by='colors', count_by='counts', style='barchart', h=80),
         rt.boxplot(df, bin_by='colors', color_by='colors', count_by='counts', style='boxplot', h=80),
         rt.boxplot(df, bin_by='colors', color_by='colors', count_by='counts', style='boxplot_w_swarm', h=80)])
```
